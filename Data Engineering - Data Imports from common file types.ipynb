### JSON, APIs

```python
# Connecting to Yelp Business Search API. In order to extract data setting up application is needed. 
api_url = "https://api.yelp.com/v3/businesses/search"
headers = {'Authorization': 'Bearer api key'} # Use API key that is dedicated to your app
params = {'term': 'cafe', 'location': 'NYC'} # Modify params to access necessary data
```

```python
import requests
import pandas as pd
```

```python
# Requesting the data from API
response = requests.get(api_url,
                       headers=headers,
                       params=params)
```

```python
# Exctracting the response
data = response.json()
```

```python
# Loading to df
cafes = pd.DataFrame(data)
```
